
# FUTURE_DS_02 - Customer Retention & Churn Analysis

## Objective
Analyze customer churn patterns, retention trends, and customer behavior for a subscription-based business.

## Tools Used
- Python
- Pandas
- Matplotlib
- Seaborn
- NumPy


```

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('customer_churn.csv')

# Display first 5 rows
df.head()
```


## Dataset Information


```

# Dataset information
df.info()
```

```

# Check missing values
df.isnull().sum()
```


## Data Cleaning


```

# Remove duplicate rows
df = df.drop_duplicates()

# Fill missing values if needed
df = df.fillna(0)

# Dataset shape
df.shape
```


## Churn Analysis


```

# Churn distribution
churn_counts = df['Churn'].value_counts()
print(churn_counts)

# Plot churn distribution
plt.figure(figsize=(6,4))
plt.bar(churn_counts.index.astype(str), churn_counts.values)
plt.xlabel('Churn')
plt.ylabel('Count')
plt.title('Customer Churn Distribution')
plt.show()
```


## Retention Rate


```

# Calculate retention rate
retention_rate = (df['Churn'] == 0).mean() * 100
print(f'Retention Rate: {retention_rate:.2f}%')
```


## Monthly Charges Analysis


```

# Average monthly charges
avg_charges = df.groupby('Churn')['MonthlyCharges'].mean()
print(avg_charges)

# Plot
avg_charges.plot(kind='bar', figsize=(6,4), title='Average Monthly Charges by Churn')
plt.ylabel('Monthly Charges')
plt.show()
```


## Customer Tenure Analysis


```

# Tenure statistics
tenure_stats = df.groupby('Churn')['Tenure'].mean()
print(tenure_stats)

# Plot
tenure_stats.plot(kind='bar', figsize=(6,4), title='Average Tenure by Churn')
plt.ylabel('Tenure')
plt.show()
```


## Key Insights

- Customers with lower tenure are more likely to churn.
- High monthly charges may increase churn probability.
- Retention strategies should focus on early-stage customers.



## Recommendations

1. Offer loyalty programs for long-term customers.
2. Provide discounts for high monthly charge plans.
3. Improve customer support for new users.
4. Monitor customer engagement regularly.
